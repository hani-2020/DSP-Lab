```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.signal import hamming, freqz
from scipy.fft import fft, ifft
```

```python
def freq_resp(b):
    w,h=freqz(b)
    plt.plot(w/max(w),abs(h))
    plt.ylabel('magnitude')
    plt.xlabel('normalized frequency')
    plt.title('frequency response')
```

```python
n=121
hpf=signal.firwin(n, cutoff=0.6, window='hamming', pass_zero=False)
freq_resp(hpf)
```

```python
t=np.arange(0,1,1/101)
f1=5
f2=20
f3=40
a1=2
a2=3
a3=4
s1=a1*np.sin(2*np.pi*f1*t)
s2=a2*np.sin(2*np.pi*f2*t)
s3=a3*np.sin(2*np.pi*f3*t)
sig=s1+s2+s3
freq_resp(sig)
sig_fft=fft(sig)
```

```python
n=101
hpf=signal.firwin(n, cutoff=0.6, window='hamming', pass_zero=False)
hpf_fft=fft(hpf)
b=ifft(hpf_fft*sig_fft)
freq_resp(b)
```
