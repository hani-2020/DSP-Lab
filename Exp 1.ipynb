```python
import numpy as np
import matplotlib.pyplot as plt
import math
n=np.arange(-5,6)
x=np.zeros_like(n)
```

```python
x[n==0]=1
plt.stem(n,x)
plt.xticks(n)
```

```python
x[(n>=0)&(n<=4)]=1
plt.stem(n,x)
plt.xticks(n)
```

```python
x=np.zeros_like(n)
x[(n>=-3)&(n<0)]=-1
x[(n>=0)&(n<=3)]=1
plt.stem(n,x)
plt.xticks(n)
```

```python
n=np.arange(-5,6)
x=np.zeros_like(n)
x[n]=6+n
x[n<=0]=0
plt.stem(n,x)
plt.xticks(n)
```

```python
a = np.arange(6)
b = np.arange(4, -1, -1)
x = np.concatenate([a, b])
plt.stem(x)
plt.xticks(np.arange(11))
```

```python
a=np.array([2, -1 , 4, 1])
plt.stem(a)
```

```python
n=np.arange(41)
a=np.exp(-((1/12)+(1j*n*np.pi/6)))
b=np.real(a)
c=np.imag(a)
fig, (ax1,ax2,ax3)=plt.subplots(nrows=3,ncols=1,figsize=(8,8))
fig.tight_layout(pad=3)
ax1.plot(n,a)
ax2.plot(n,b)
ax3.plot(n,c)
```

```python
n=np.arange(51)
a=(0.95**n)*(np.cos(0.1*np.pi*n))
plt.plot(n,a)
```

```python
fig, (ax1,ax2,ax3,ax4,ax5)=plt.subplots(5,1,figsize=(15,15))
fig.tight_layout(pad=3)
a=20*(0.9**n)
b=0.2*(1.2**n)
c=-0.8**n
d=-4*(0.8**n)
e=2*n*(0.9**n)
ax1.stem(n,a)
ax2.stem(n,b)
ax3.stem(n,c)
ax4.stem(n,d)
ax5.stem(n,e)
```

```python
n=np.arange(21)
x1=np.sin(0.2*np.pi*n)
x2=np.sin(1.8*np.pi*n)
x3=np.sin(2.2*np.pi*n)
fig, (ax1,ax2,ax3)=plt.subplots(3,1,figsize=(8,8))
fig.tight_layout(pad=3)
ax1.stem(n,x1)
ax2.stem(n,x2)
ax3.stem(n,x3)
```

```python
t = np.arange(0, 1, .02)
x = np.sin(2*np.pi*1*t)
y = np.sin(2*np.pi*2*t)
plt.plot(t, x, label = '1 Hz')
plt.plot(t, y, label = '2 Hz')
plt.legend()
```

```python
t=np.arange(-2*np.pi,2*np.pi,0.01)
y=np.sin(t)
z=abs(y)
fig, (ax1,ax2)=plt.subplots(2,1,figsize=(8,8))
fig.tight_layout(pad=3)
ax1.plot(t,y)
ax2.plot(t,z)
```

```python
t = np.arange(0, 2, .01)
signal = np.sin(2*np.pi*t)
noise1 = np.random.normal(0, 0.1, t.size)
noisy_signal1 = signal + noise1
noise2 = np.random.normal(0, 0, t.size)
noisy_signal2 = signal + noise2
noise3 = np.random.normal(0, 1, t.size)
noisy_signal3 = signal + noise3
noise4 = np.random.normal(0, 2, t.size)
noisy_signal4 = signal + noise4
fig, (ax1,ax2,ax3,ax4)=plt.subplots(4,1,figsize=(8,8))
fig.tight_layout(pad=3)
ax1.plot(noisy_signal1)
ax2.plot(noisy_signal2)
ax3.plot(noisy_signal3)
ax4.plot(noisy_signal4)
```
